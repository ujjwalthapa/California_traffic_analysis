```python
# Create a new DataFrame that looks into SF county
Sf_df = ca_dmv_peak_hours.loc[ca_dmv_peak_hours["COUNTY"] == "SF"]
Sf_df
```

```python
Sf_df.plot(kind="line", figsize=(15,3))

# Set a title for the chart
plt.title("Average Cars in SF County")

#Add the legend.
legend = plt.legend(loc = "right")

legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]


plt.show()
plt.tight_layout()
```

```python
# SF DAY (AM_SUMMARY)
#print AM_K_FACTOR_AMT
Am_summary = (round(Sf_df["AM_PEAK_VOL_PEAK_DIR"].sum()))
Am_summary
```

```python
#print AM_D_FACTOR_AMT
Total_AM_K = (round(Sf_df["AM_K_FACTOR_AMT"].sum()))
Total_AM_K 
```

```python
#print AM_KD_FACTOR
Total_AM_D  = (round(Sf_df["AM_D_FACTOR_AMT"].sum()))
Total_AM_D
```

```python
SF_type = Sf_df.set_index('COUNTY')['ROUTE']

SF_type.value_counts()
```

```python
#Create DataFrame Display
SF_AM = pd.DataFrame({"AM_PEAK_VOL_PEAK_DIR": Am_summary,
                 "AM_K_FACTOR_AMT": Total_AM_K ,
                 "AM_D_FACTOR_AMT": Total_AM_D,
                 "ROUTE": SF_type
})

SF_AM.head()
```

```python
# Calculate Type Percents
amcity_fare = Sf_df.groupby(["ROUTE"])["AM_PEAK_VOL_PEAK_DIR"].sum()

amcity_fare
```

```python
#Use the city_fare index for wedge labels.
SF_type.index
```

```python
amcity_fare.plot(kind="bar", figsize=(15,3))

# Set a title for the chart
plt.title("Busiest AM/PM Route in SF")

#Add the legend.
legend = plt.legend(loc = "upper left")


plt.show()
plt.tight_layout()
```
